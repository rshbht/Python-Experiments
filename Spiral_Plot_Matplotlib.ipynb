```python
import matplotlib.pyplot as plt
import numpy as np

# Create an array of values for the angle
t = np.arange(0, 10*np.pi, 0.1)

# Create the x and y coordinates using the angle and a scaling factor
x = t * np.cos(t)
y = t * np.sin(t)

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the spiral
ax.plot(x, y)

# Show the plot
plt.show()
```
