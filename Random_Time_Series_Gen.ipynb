```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate some sample data
dates = pd.date_range('2022-01-01', periods=365)
values = np.random.randn(365).cumsum()

# Create a new figure
fig, ax = plt.subplots()

# Plot the data as a time series
ax.plot(dates, values)

# Format the plot
ax.set(xlabel='Date', ylabel='Value', title='Time Series Graph')
ax.grid()

# Show the plot
plt.show()
```
